```python
import cv2 

# Create an object to read  
# from camera 
video = cv2.VideoCapture('../archive/SeaTurtle1/SeaTurtle1.mp4') 
   
# We need to check if camera 
# is opened previously or not 
if (video.isOpened() == False):  
    print("Error reading video file") 

# We need to set resolutions. 
# so, convert them from float to integer. 
frame_width = int(video.get(3)) 
frame_height = int(video.get(4)) 
   
size = (frame_width, frame_height) 
   
# Below VideoWriter object will create 
# a frame of above defined The output  
# is stored in 'filename.avi' file. 
result = cv2.VideoWriter('filename.avi',  
                         cv2.VideoWriter_fourcc(*'XVID'), 
                         10, size) 
    
while(True): 
    ret, frame = video.read() 
  
    if ret == True:  
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        blur = cv2.GaussianBlur(gray, (3, 3), 0)
        canny = cv2.Canny(blur, 10, 70)
        ret, mask = cv2.threshold(canny, 70, 255, cv2.THRESH_BINARY)
  
        # Write the frame into the 
        # file 'filename.avi'

        result.write(mask) 
  
        # Display the frame 
        # saved in the file 
        cv2.imshow('Frame', mask) 
  
        # Press S on keyboard  
        # to stop the process 
        if cv2.waitKey(25) & 0xFF == 27: 
            break
  
    # Break the loop 
    else: 
        break

# When everything done, release  
# the video capture and video  
# write objects 
video.release() 
result.release() 
    
# Closes all the frames 
cv2.destroyAllWindows() 
```

```python
import cv2 
import numpy as np

# Create an object to read  
# from camera 
video = cv2.VideoCapture('../archive/SeaTurtle1/SeaTurtle1.mp4') 
   
# We need to check if camera 
# is opened previously or not 
if (video.isOpened() == False):  
    print("Error reading video file") 

# We need to set resolutions. 
# so, convert them from float to integer. 
frame_width = int(video.get(3)) 
frame_height = int(video.get(4)) 
   
size = (frame_width, frame_height) 
   
while(True): 
    ret, frame = video.read() 
  
    if ret == True:  
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
       
        blur = cv2.GaussianBlur(gray, (3, 3), 0)
        
        sobelx = cv2.Sobel(blur,cv2.CV_64F,1,0,ksize=5)  # x
        sobely = cv2.Sobel(blur,cv2.CV_64F,0,1,ksize=5)  # y
        
        gradient_magnitude = np.sqrt(np.square(sobelx) + np.square(sobely))
        
        ret, mask = cv2.threshold(gradient_magnitude, 2500, 7172.6, cv2.THRESH_BINARY)
  
        # Display the frame 
        # saved in the file 
        cv2.imshow('Frame', mask) 
  
        # Press S on keyboard  
        # to stop the process 
        if cv2.waitKey(1) & 0xFF == 27: 
            break
  
    # Break the loop 
    else: 
        break

# When everything done, release  
# the video capture and video  
# write objects 
video.release() 
result.release() 
    
# Closes all the frames 
cv2.destroyAllWindows() 
```
